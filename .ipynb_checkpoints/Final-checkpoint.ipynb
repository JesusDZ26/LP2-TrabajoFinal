```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
def informacion(entry):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.128 Safari/537.36'}

    url = f"https://www.uniprot.org/uniprot/?query={entry}+Bos+taurus&sort=score" 
    r = requests.get(url, headers=headers) # url general:varía para cada gen entonces hacemos uso del "tag", y por ende cada gen abreviado redirecciona a su propia página 

    soup3 =BeautifulSoup(r.text, "html.parser") #lee el texto html de la pagina donde se encuentra el entry y entry_name de cada gen
    DUDA=soup3.find_all('div',{"id":"noResultsMessage"})
        
    if DUDA == []:   #Si no hay un mensaje de información no encontrada que busque el entry ,el link y el entry name 
        soup =BeautifulSoup(r.text, "html.parser")
        questions =soup.find_all('tr', class_='entry selected-row') #Busca la tabla donde se encuentran las variables entry y entryname
        for item in questions: #Bucle que busca las ubicaciones del entry,link y entryname para cada gen 
            Entry = item.find('td',{'class':'entryID'}).text
            link='https://www.uniprot.org' +item.find('td', {'class': 'entryID'}).a['href']
            Entry_names = item.find('td',{'class':''}).text

        url2= f"https://www.uniprot.org/uniprot/{Entry}" #Redirecciona la página donde se encuentra GBIO,MOLECULAR-F,BIOLOGICAL-P
        r2 = requests.get(url2, headers=headers)
        soup2 =BeautifulSoup(r2.text, "html.parser")

        questions2 =soup2.find_all('div', {'class':'content entry_view_content up_entry swissprot'}) #Busca la tabla donde se encuntra
        #las variables gen,protein,organismo y el status 

        for item in questions2:
            Protein = item.find('div',{'id':'content-protein'}).text
            Gene=item.find('div',{'id':'content-gene'}).text
            Organism=item.find('div',{'id':'content-organism'}).text
            Status=item.find('a',{'title':'Reviewed (Swiss-Prot)'}).text
        
        try: #Almacena la información si es que se encontraron las variables
            #print(Protein, Gene, Organism, Status) 
            datos1 = [Protein, Gene, Organism, Status]
        except UnboundLocalError: #En caso no se haya encontrado la tabla,devuelve para cada variable Not found(no encontrado)
            Protein = "Not found Protein"; Gene = "Not found Gene"
            Organism = "Not found Organism"; Status = "Not found Status" 
            #datos1 = [Protein, Gene, Organism, Status]
            
        go2=soup2.find("ul",{"class":"noNumbering molecular_function"}) #Busca dentro del body la sección en que se encuentra 
        #el molecular function
        if go2 == None: #Si la clase noNumbering molecular_function no es encontrada,que imprima vacío
            print(" ")
        else :
            goes=go2.find_all("li") #Caso contrario que se ubique en las líneas de información de molecular_function

        cadena2=""
        try:   #Al ser encontrada la información de GoBiologicalprocess,que recorra cada linea de texto 
            for j in goes:  
                Molecular_function=str(j.find("a").text).capitalize()  #imprime todo texto de bio_process con las palabras iniciales en mayúscusla  
                cadena3=Molecular_function+"."
                cadena2=cadena2+cadena3 #La variable almacena toda la información en un párrafo
        except UnboundLocalError:  #Caso contrario,que imprima cadena no encontrada
            cadena2 = "Cadena2 nor found " #

        go=soup2.find("ul",{"class":"noNumbering biological_process"}) #Busca dentro del body la sección en que se encuntra 
        #el Gobiologicalprocess
        if go == None:  #En caso no sea encontrada que reporte vacio 
            print(" ")
        else :  #De lo contrario que busque las líneas que contengan información
            goes=go.find_all("li")

        cadena=""  #Variable en que se almacenará la información
        try:  #Si se ah encontrado la clase noNumbering biological_process ,que recorra cada línea de información 
            for j in goes:
                Biological_process_row=str(j.find("a").text).capitalize()  #imprime todo texto de bio_process   
                cadena1=Biological_process_row+"."  
                cadena=cadena+cadena1 #Que almacene información del biological_process
        except UnboundLocalError:  #Si no se encontró ,que imprima cadena not found
            cadena = "Cadena1 not found"

        #KEYWORDS:"Molecular function', 'Biological process"
        linea1=[]; linea2=[]; linea3=[]  #Creamos 3 listas vacías 
        #Lista 1, mostrará los datos que están primera columna de la tabla Keywords,tabla no mayor de 2x2
        #Lista 2, muestra los datos que están en la segunda columna de la tabla (la información requerida),tabla de dimensión no mayor de 2x2
        #Lista 3, muestra los datos que estan en la tercera columna de la tabla ,tabla con dimensión mayor a 2x2
        division=soup2.find("div", {"id":"function"} ) #Para un búqueda más rápida,usamos el id(se ubica la tabla)

        try: #Intenta buscar la tabla 
            tabla=division.find("table", {"class":"databaseTable"})
            rows=tabla.find_all("tr")
        except AttributeError: #En caso no encontrado la tabla,imprime not found
            molecular_f="Not found molecular_f"
            biological_p="Not found biological_p"
        else:  #Si en caso ya se ah encontrado la tabla, que recorra cada fila de la tabla
            for row in rows:
                linea1.append(str(row.find_all("td")[0].text))   #muestra ('Molecular function', 'Biological process')
                linea2.append(str(row.find_all("td")[1].text))   #muestra ("'Activator, DNA-binding','Transcription, Transcription regulation'")
                try: #Si la tabla es de mayor dimensión de 2x2, almacenará información de la columna 2 de la tabla
                    linea3.append(str(row.find_all("td")[1].a.text))
                except AttributeError: #Si en caso no se haya encontrado una tabla cuya dimensión no se mayor de 2x2 que continue 
                    #con el siguiente código
                    pass
            if len(linea1)==1: #Si la longitud de la linea 1 es equivalente a 1(linea1=lista)
                try: #Intenta almacenar la información en la lista linea3
                    linea3.append(str(row.find_all("td")[1].a.text)) 
                except AttributeError: #Si en caso manifiesta error, quiere decir que no es una tabla de extensión 1x2
                    if len(linea3)!=0: #Si la longitud de la linea 3 es diferente a cero(tabla con una o más de una fila)
                        if linea1[0]=="Molecular function": #Si el elemento de la posición de la línea 1 es equivalente a M.function
                            #entonces almacenará el primer elemento en la línea 3
                            molecular_f = linea3[0]
                            biological_p="Not found biological_p"
                        if linea1[0]=="Biological process": #Si el elemento incial de la linea 1 es equivalente a B.process 
                            #entonces la información del primer elemento de la línea 3, se almacena en la variables biological_p
                            molecular_f="Not found biological_p"
                            biological_p= linea3[0]
                    else:
                        molecular_f="Not found molecular_f" #Se imprime el no encontrado
                        biological_p="Not found biological_p"
            if len(linea1)>=2:  #Si la longitud de la línea 1 es mayor o igual a 2, quiere decir que la tabla tiene de 2 a más filas
                molecular_f=linea2[0] #Almacena la información del elemento 1 de la línea 2
                biological_p=linea2[1] #Almacena la información del elemento2 de la línea 2
                #print(molecular_f,biological_p)
            if len(linea1)==0: #Si en caso la longitud de la línea es igual a 0, no almacenará cada información e imprime not found
                #para cada variable 
                molecular_f="Not found molecular_f"
                biological_p="Not found biological_p"
        
        try: #Almacena todas las variables ya mencionadas en una lista denominada datos

            datos = [Entry, Entry_names,Protein, Gene,Organism,Status,cadena2+cadena, molecular_f, biological_p]
        except UnboundLocalError: #En caso encuentra algún error en la sección de los Keywords, que imprima elementos no encontrados 
            molecular_f="Not found molecular_f"
            biological_p="Not found biological_p"
            datos = [Entry, Entry_names,Protein, Gene,Organism,Status,cadena2+cadena, molecular_f, biological_p]
    else: #Si el gen ingresado no tiene información del entry o entry_name, entonces no se rediccionará a la siguiente página
        #donde se encuentra la información de las otras variables
        datos = ["Gene not found"]*9  #Se imprimirá gene not found para cada variable almacenada en datos 
    # datos = [Entry, Entry_names,Protein, Gene,Organism,Status,cadena2+cadena, molecular_f, biological_p]
    return(datos) #Devuelve las informaciones que contiene datos 
#informacion("CD47")
```
